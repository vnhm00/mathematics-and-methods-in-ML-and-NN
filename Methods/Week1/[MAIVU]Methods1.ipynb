## 3. Load CKD dataset

```python
X = pd.read_csv('kd.csv', na_values=['?']).fillna(float('nan')) #Read file and replace missing value with NaN
X.sample(10) #take 10 random rows from X
```

```python
X_affected = X[~X['class'].str.contains("notckd")] #filter out those are not affected
X_affected.sample(10)
```

```python
X_affected.describe() #calculate the mean, std, min, max, etc.
```

```python
X_affected.corr() #print correlation matrix of numerical variables
```

```python
#visualize the correlation matrix
import seaborn as sns
plt.figure(figsize = (10,8))
corr = X_affected.corr()
ax = sns.heatmap(
    corr, 
    vmin=-1, vmax=1, center= 0,
    cmap=sns.diverging_palette(20, 220, n=200),
    square=True
)
ax.set_xticklabels(
    ax.get_xticklabels(),
    rotation=45
);
```
